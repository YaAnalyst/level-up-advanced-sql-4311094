# Try to use scatter plot to explore the dependencies between features!

```python
df.Quantity.describe(percentiles=[0.99])
```

```python
# Correlation of Quantity and UnitPrice

df[(df.UnitPrice>=0) & (df.UnitPrice<=18) & (df.Quantity>=0) & (df.Quantity<=100)].plot.scatter(x='UnitPrice', y='Quantity');
```

```python
# Correlationt of TotalValue and UnitPrice in France

df[(df.UnitPrice>=0) & (df.UnitPrice<=18) & (df.Quantity>=0) & (df.Quantity<=100) & (df.Country =='France')].plot.scatter(x='UnitPrice', y='TotalValue');
```

```python
# Correlationt of TotalValue and UnitPrice in Czech Republic

df[(df.UnitPrice>=0) & (df.UnitPrice<=18) & (df.Quantity>=0) & (df.Quantity<=100) & (df.Country =='Czech Republic')].plot.scatter(x='UnitPrice', y='TotalValue');
```
